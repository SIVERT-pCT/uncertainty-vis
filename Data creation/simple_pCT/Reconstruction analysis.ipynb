```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
num_angle = 101
num_offset = 1
num_spotx = 130
chord_length = True
RSP_shape = [190, 1226]
filter_name='ramp'
chords = 'exact' if chord_length else 'map'
```

```python
mean1 = np.load('../../Data/simple_pCT/Reconstruction/Head/{:d}_{:d}/RSP_angles{:d}_offset{:d}_spotx{:d}_{:s}_{:s}.npy'.format(RSP_shape[0], RSP_shape[1], num_angle, num_offset, num_spotx, chords, filter_name))
Sigma1 = np.load('../../Data/simple_pCT/Sigma/Sigma_raedler_angles{:d}_offset{:d}_spotx{:d}_{:s}_{:s}_{:d}_{:d}.npy'.format(num_angle, num_offset, num_spotx, chords, filter_name, RSP_shape[0], RSP_shape[1]))
var1 = np.abs(np.diag(Sigma1)).reshape((num_spotx,num_spotx))

print(mean1.shape)
print(Sigma1.shape)
```

```python
plt.figure(figsize=(18,10))
plt.subplot(1,2,1)
plt.title(r'$\mu_1$')
plt.imshow(mean1, aspect='auto')
plt.colorbar()
plt.subplot(1,2,2)
plt.title(r'$\sigma_1^2$')
plt.imshow(var1)
plt.colorbar()
plt.show()
```

```python
num_angle = 180
num_offset = 1
num_spotx = 130
chord_length = True
RSP_shape = [190, 1226]
filter_name='cosine'
chords = 'exact' if chord_length else 'map'
```

```python
mean2 = np.load('../../Data/simple_pCT/Reconstruction/Head/{:d}_{:d}/RSP_angles{:d}_offset{:d}_spotx{:d}_{:s}_{:s}.npy'.format(RSP_shape[0], RSP_shape[1], num_angle, num_offset, num_spotx, chords, filter_name))
Sigma2 = np.load('../../Data/simple_pCT/Sigma/Sigma_raedler_angles{:d}_offset{:d}_spotx{:d}_{:s}_{:s}_{:d}_{:d}.npy'.format(num_angle, num_offset, num_spotx, chords, filter_name, RSP_shape[0], RSP_shape[1]))
var2 = np.abs(np.diag(Sigma2)).reshape((num_spotx,num_spotx))

print(mean2.shape)
print(Sigma2.shape)
```

```python
plt.figure(figsize=(18,10))
plt.subplot(1,2,1)
plt.title(r'$\mu_2$')
plt.imshow(mean2, aspect='auto')
plt.colorbar()
plt.subplot(1,2,2)
plt.title(r'$\sigma_2^2$')
plt.imshow(var2)
plt.colorbar()
plt.show()
```

```python
vmin = min(mean1.min(), mean2.min())
vmax = max(mean1.max(), mean2.max())

plt.figure(figsize=(18,10))
plt.subplot(1,2,1)
plt.imshow(mean1, aspect='auto', vmin=vmin, vmax=vmax)

plt.subplot(1,2,2)
plt.imshow(mean2, aspect='auto', vmin=vmin, vmax=vmax)
plt.show()

plt.figure(figsize=(18,10))
plt.imshow(mean1 - mean2, aspect='auto', cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
vmin = min(var1.min(), var2.min())
vmax = max(var1.max(), var2.max())

plt.figure(figsize=(18,10))
plt.subplot(1,2,1)
plt.imshow(var1, aspect='auto', vmin=vmin, vmax=vmax)

plt.subplot(1,2,2)
plt.imshow(var2, aspect='auto', vmin=vmin, vmax=vmax)
plt.show()

plt.figure(figsize=(18,10))
plt.imshow(var1 - var2, aspect='auto', cmap='RdBu')
plt.colorbar()
plt.show()
```
