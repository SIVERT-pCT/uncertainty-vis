```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def normal_dist(x , mean , var):
    prob_density = 1 / (np.sqrt(2*np.pi)*var) * np.exp(-0.5*((x-mean)/var)**2)
    return prob_density
```

```python
mu_x1 = 5
sigma_x1 = 1
mu_x2 = 1.5
sigma_x2 = 0.5

mu_y1 = 4.5
sigma_y1 = 0.5
mu_y2 = 5
sigma_y2 = 1

T_v = 100
v1 = np.linspace(1, 8, T_v)

x1 = normal_dist(v1, mu_x1, sigma_x1**2)
y1 = normal_dist(v1, mu_y1, sigma_y1**2)

z1 = x1 * np.log(1 / y1) - x1 * np.log(1 / x1)

plt.figure(figsize=(18,8))
plt.subplot(3,1,1)
plt.title(r'$x_1$')
plt.plot(v1, x1)
plt.grid()

plt.subplot(3,1,2)
plt.title(r'$y_1$')
plt.plot(v1, y1)
plt.grid()

plt.subplot(3,1,3)
plt.title(r'$z_1 = x_1(v) \log \left( \frac{x_1(v)}{y_1(v)} \right)$')
plt.fill_between(v1, z1)
plt.hlines(0, 0, 8, colors='k')
plt.grid()

plt.show()
```
