```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def normal_dist(x , mean , var):
    return 1 / np.sqrt(2*np.pi * var) * np.exp(-0.5*(x-mean)**2/var)
```

```python
z1 = np.random.default_rng()
z2 = np.random.default_rng()
```

```python
mu1 = -3
sigma1 = 1.5
mu2 = 5
sigma2 = 0.75

samples1 = z1.normal(mu1, sigma1, 100000)
samples2 = z2.normal(mu2, sigma2, 100000)
samples3 = samples1 + samples2
```

```python
y1 = normal_dist(samples1, mu1, sigma1) * z1.normal(0, 1, 100000)
y2 = normal_dist(samples2, mu2, sigma2) * z2.normal(0, 1, 100000)
y3 = normal_dist(samples3, mu1+mu2, np.sqrt(sigma1**2 + sigma2**2)) * z2.normal(0, 1, 100000)

mu3 = np.mean(samples3)
sigma3 = np.std(samples3)

print(mu3, mu1 + mu2)
print(sigma3, np.sqrt(sigma1**2 + sigma2**2))

plt.figure(figsize=(18,6))
plt.scatter(samples1, y1, c='r', label=r'$z_1$')
plt.scatter(samples2, y2, c='b', label=r'$z_2$')
plt.scatter(samples3, y3, c='k', label=r'$z_1 + z_2$')
plt.legend()
plt.grid()
plt.show()
```
