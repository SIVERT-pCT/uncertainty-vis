```python
import numpy as np
from matplotlib import pyplot as plt
import tensorflow as tf
```

```python
x = tf.Variable([np.linspace(0,np.pi)])

with tf.GradientTape() as tape:
    tape.watch(x)
    
    y = tf.sin(x)
    
dy_dx = tape.gradient(y, x)

plt.plot(x[0], y[0])
plt.plot(x[0], dy_dx[0])
plt.show()
```
