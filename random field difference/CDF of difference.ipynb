```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def normal_dist(x , mean , var):
    prob_density = 1 / (np.sqrt(2*np.pi)*var) * np.exp(-0.5*((x-mean)/var)**2)
    return prob_density
def wasserstein(mu1, var1, mu2, var2):
    return np.sqrt(
        ( mu1 - mu2 )**2 + ( np.sqrt(var1) - np.sqrt(var2) )**2
    )
```

```python
mu_x1 = 5
sigma_x1 = 1
mu_x2 = 1.5
sigma_x2 = 0.5

mu_y1 = 4.5
sigma_y1 = 0.5
mu_y2 = 5
sigma_y2 = 1

T_v = 100
v1 = np.linspace(1, 8, T_v)

cdf_x1 = np.cumsum(normal_dist(v1, mu_x1, sigma_x1**2))
cdf_x1 /= cdf_x1.max()
cdf_y1 = np.cumsum(normal_dist(v1, mu_y1, sigma_y1**2))
cdf_y1 /= cdf_y1.max()

cdf_z1 = (cdf_x1 - cdf_y1)**2

plt.figure(figsize=(18,8))
plt.subplot(3,1,1)
plt.title(r'$F_1$')
plt.plot(v1, cdf_x1)
plt.grid()

plt.subplot(3,1,2)
plt.title(r'$F_1$')
plt.plot(v1, cdf_y1)
plt.grid()

plt.subplot(3,1,3)
plt.title(r'$z_1 = (F_1(v) - F_2(v))^2$')
plt.fill_between(v1, cdf_z1)
plt.grid()

plt.show()
```
